```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import rcParams
from sklearn.cluster import DBSCAN
```

```python
rcParams['figure.figsize'] = 15, 10
sns.set_style('darkgrid')
```

```python
customer_df = pd.read_csv('customers.csv')
customer_df.drop(['CustomerID'], axis=1, inplace=True)
customer_df.head()
```

```python
sns.heatmap(customer_df.corr(), annot=True)
plt.show()
```

```python
plt.pie(customer_df['Genre'].value_counts(), labels=['Female','Male'], colors=['Pink','Blue'], shadow=True,
        explode=[0,0.1])
plt.legend()
plt.axis('equal')
plt.tight_layout()
plt.show()
```

```python
# Saving only Spending_Score and income values into X.
X = customer_df.iloc[:, [2, 3]].values
```

```python
##eps = epsilon , min_samples = minpts, distance =euclidean
db = DBSCAN(eps=3, min_samples=4, metric='euclidean')
db.fit(X)
db.labels_
```

```python
y_preds = db.fit_predict(X)
```

```python
plt.scatter(X[y_preds == 0, 0], X[y_preds == 0, 1], s = 80, c = 'pink', label='C1')
plt.scatter(X[y_preds == 1, 0], X[y_preds == 1, 1], s = 80, c = 'orange', label='cluster 2')
plt.scatter(X[y_preds == 2, 0], X[y_preds == 2, 1], s = 80, c = 'yellow', label='cluster 3')
plt.scatter(X[y_preds == 3, 0], X[y_preds == 3, 1], s = 80, c = 'lightgreen', label='cluster 4')
plt.scatter(X[y_preds == 4, 0], X[y_preds == 4, 1], s = 80, c = 'lightblue', label='cluster 5')
plt.scatter(X[y_preds == 5, 0], X[y_preds == 5, 1], s = 80, c = 'indigo', label='cluster 6')
plt.scatter(X[y_preds == 6, 0], X[y_preds == 6, 1], s = 80, c = 'purple', label='cluster 7')
plt.scatter(X[y_preds == 7, 0], X[y_preds == 7, 1], s = 80, c = 'black', label='cluster 8')
plt.scatter(X[y_preds == 8, 0], X[y_preds == 8, 1], s = 80, c = 'brown', label='cluster 9')
plt.title('Clusters of customers')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.legend()
plt.show()
```
